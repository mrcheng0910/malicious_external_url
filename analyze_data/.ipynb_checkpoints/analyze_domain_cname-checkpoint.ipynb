### 分析非法网站的cname情况

* cname使用情况
* 使用cdn的情况
* 将为支持非法域名的cname（cdn）列入到黑名单中，评分机制。

说明：
含有cname的非法域名，可能其还有其他域名指向该cname，方便更换IP地址，其他域名利用cname也同时指向更换后的IP



```python
# encoding:utf-8
```

```python
from collections import Counter,defaultdict
from pymongo import MongoClient
from datetime import datetime
```

```python
% matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
client = MongoClient('172.29.152.152', 27017)
db = client['eds_last']
col = db['domain_ip_cname']
gam_records = col.find({'dm_type':'Gamble','dm_cname.cnames':{'$not': {'$size': 0}}})
por_records = col.find({'dm_type':'Porno','dm_cname.cnames':{'$not': {'$size': 0}}})
```

```python
gam_cname_counter = Counter()   # 统计含有cname的级数
gam_cname_sets = []   # cname集合，即部分域名有多个cname
gam_cname_all = []   # 所有cname
gam_domains_total = gam_records.count()

gam_cname_domain = defaultdict(set)

por_cname_counter = Counter()   # 统计含有cname的级数
por_cname_sets = []   # cname集合，即部分域名有多个cname
por_cname_all = []   # 所有cname
por_domains_total = por_records.count()


## 统计gamble域名的情况
for r in gam_records:
    if len(r['dm_cname']) >1:      # 有更新CNAME的域名
        print r['domain'], ',', len(r['dm_cname'])
    
    for c in r['dm_cname']:
        gam_cname_domain[tuple(c['cnames'])].add(r['domain'])
        gam_cname_sets.append(tuple(c['cnames']))
        gam_cname_all.extend(c['cnames'])
        gam_cname_counter[len(c['cnames'])] += 1
        
print '---------------------------'
## 统计porno域名的情况
for r in por_records:
    if len(r['dm_cname']) >1:
        print r['domain'],',',len(r['dm_cname'])
    for c in r['dm_cname']:
        por_cname_sets.append(tuple(c['cnames']))
        por_cname_all.extend(c['cnames'])
        por_cname_counter[len(c['cnames'])] += 1
        
        
```

```python
## 打印gamble
print gam_domains_total
print gam_cname_counter
print len(gam_cname_sets)
print len(set(gam_cname_sets))
print len(gam_cname_all)
print len(set(gam_cname_all))

print "-----------------"
## 打印porno
print por_domains_total
print por_cname_counter
print len(por_cname_sets)
print len(set(por_cname_sets))
print len(por_cname_all)
print len(set(por_cname_all))

```

## 说明
部分域名，例如([u'77777yyy.com', u'77777zzz.com'])，yyy和zzz是有规律的，经测试xxx，也是非法域名，所以根据这种方法也可以挖掘出恶意域名

```python
# 使用相同cname的域名集合
for i in gam_cname_domain:
    print i, gam_cname_domain[i]
```
