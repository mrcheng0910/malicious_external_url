### 分析网站使用权威DNS（name server）的情况

1. 非法网站使用ns的数量分布情况（例如2，3个ns服务器）
2. 非法网站最常使用的ns服务器统计
3. 非法网站使用的ns服务器的情况

```python
# encoding:utf-8
from collections import Counter,defaultdict
from pymongo import MongoClient
from datetime import datetime
import tldextract
% matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
client = MongoClient('172.29.152.152', 27017)
db = client['eds_last']
col = db['domain_whois']
```

```python
gam_records = col.find({})
null_ns_count = 0    # NS记录为空的域名数量
diff_ns_cnt = 0   # 含有不同ns记录（主域名）的域名数量
domain_ns_num = Counter()   # 域名含有NS记录数量
ns_domain = defaultdict(list)
ns_domain_cnt = {}
main_ns = defaultdict(list)
ns_main_domain = defaultdict(set)    # ns记录的主域名和子域名
ns_main_domain_single = defaultdict(set)

for i in gam_records:
    main_ns = set()
    main_domain = ""
    dm_name_server = i['name_server'].lower()
    
    if dm_name_server != "":
        dm_ns_list = dm_name_server.split(';')
        dm_ns_list.sort()  # 注意排序
        ns_domain[tuple(dm_ns_list)].append(i['domain'])  # 每个ns集对应的域名
        
        for j in dm_ns_list:
            r = tldextract.extract(j)
            main_domain = r.domain+'.'+r.suffix
            main_ns.add(main_domain)

        if len(main_ns)>1:   # ns服务器的主域名不同
            diff_ns_cnt += 1
        else:
            ns_main_domain[main_domain].add(tuple(dm_ns_list))
            
            for n in dm_ns_list:
                ns_main_domain_single[main_domain].add(n)


ns_main_domain_cnt = {}

# 得到每个ns的主域名对应的域名数量
for k in ns_main_domain:
    total_domain_cnt = 0
    for v in ns_main_domain[k]:
            total_domain_cnt += len(ns_domain[v]) 
    ns_main_domain_cnt[k] =  total_domain_cnt
        

# for i in ns_main_domain_cnt:
#     print i, ns_main_domain_cnt[i]
ns_main_domain_cnt= sorted(ns_main_domain_cnt.iteritems(), key=lambda d:d[1], reverse = True)  # 排序

ns_name = []
sub_ns_name = []
sub_ns_name_sets = []
sub_ns_cnt = []
ns_domain = []

for n,c in ns_main_domain_cnt:
    temple_ns = set()
    ns_name.append(n)
    ns_domain.append(c)
    sub_ns_name.append(list(ns_main_domain_single[n]))
    sub_ns_cnt.append(len(ns_main_domain[n]))
    sub_ns_name_sets.append(ns_main_domain[n])


## 制表
# header = ['name server','sub name server','sub name server','domain numbers']
from tabulate import tabulate
from IPython.display import HTML, display
n = 15
table = {
    "name server": ns_name[:n],
    "sub_ns_name": sub_ns_name[:n],
    "sub_ns_cnt":sub_ns_cnt[:n],
    "ns_domain": ns_domain[:n],
    "sub_ns_name_set":sub_ns_name_sets[:n]
}
display(HTML(tabulate(table, tablefmt='html',headers='keys')))   
```
